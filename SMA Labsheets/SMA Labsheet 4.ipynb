##### Apply transformations to the company features by removing the company related suffixes

```python
def remove_company_suffix(org):
    if pd.notna(org):
        suffixes = ["Inc.", "Ltd.", "Corp.", "LLC", "Co.", "AG", "GmbH", "BV", "S.A.", "Pvt. Ltd.", "LLP"]
        for suffix in suffixes:
            org = org.replace(suffix, "").strip()
    return org
```

```python
df['Organization']
```

```python
df['Organization'] = df['Organization'].apply(remove_company_suffix)
df['Organization']
```

```python
from opencage.geocoder import OpenCageGeocode #pip install OpenCage
import folium
```

```python
def geocode_company(org): #retrieving geocode
    api_key = "feecb3f7edc84718a351e579606425e8"
    geocoder = OpenCageGeocode(api_key)
    try:
        result = geocoder.geocode(org)
        if result and len(result):
            latitude = result[0]['geometry']['lat']
            longitude = result[0]['geometry']['lng']
            return latitude, longitude
    except Exception as e:
        print(f"Error geocoding {org}: {e}")

    return None, None
```

```python
df['Latitude'], df['Longitude'] = zip(*df['Organization'].apply(geocode_company))
print(df)
```

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
connections_map = folium.Map(location=map_center, zoom_start=10)

for _, row in df.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=row['Name'],
        ).add_to(connections_map)
connections_map.save("connections_map.html")
```

```python
import matplotlib.pyplot as plt #plotting for the respective geocode
plt.figure(figsize=(12, 10))

plt.scatter(df['Longitude'], df['Latitude'], marker='o', alpha=0.7)

for _, row in df.iterrows():
    plt.annotate(row['Name'], (row['Longitude'], row['Latitude']), fontsize=8)

plt.title("LinkedIn Connections")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.grid(True)
plt.show()
```
