##### Step 4: Find Jaccard distance between two titles

##### Step 5: Calculate the distance and similarity between 2 titles

```python
from nltk.metrics.distance import jaccard_distance 

job_title_1 = 'Chief Executive Officer'.split()
job_title_2 = 'Chief Technology Officer'.split()

print(job_title_1)
print(job_title_2)

print()
print('Intersection:')
intersection = set(job_title_1).intersection(set(job_title_2))
print(intersection)

print()
print('Union:')
union = set(job_title_1).union(set(job_title_2))
print(union)

print()
print('Similarity:', len(intersection) / len(union))
print('Distance:', jaccard_distance(set(job_title_1), set(job_title_2)))
```

```python
import pandas as pd
from nltk.util import bigrams
from nltk.metrics.distance import jaccard_distance

# Sample job titles for demonstration
job_titles = [
    'Chief Executive Officer',
    'Chief Technology Officer',
    'Vice President, Sales',
    'Vice President, Customer Relations',
    'Software Engineer',
    'Data Scientist',
    'Product Manager'
]

# Calculate Jaccard distance between job titles
def jaccard_similarity(title1, title2):
    title1_bigrams = list(bigrams(title1.split(), pad_left=True, pad_right=True))
    title2_bigrams = list(bigrams(title2.split(), pad_left=True, pad_right=True))
    return 1 - jaccard_distance(set(title1_bigrams), set(title2_bigrams))

# Create a similarity threshold
similarity_threshold = 0.5

# Create a dictionary to store clusters
clusters = {}

# Iterate through job titles and build clusters
for i, title1 in enumerate(job_titles):
    cluster_found = False
    for cluster_id, cluster in clusters.items():
        for title2 in cluster:
            similarity = jaccard_similarity(title1, title2)
            if similarity >= similarity_threshold:
                clusters[cluster_id].append(title1)
                cluster_found = True
                break
        if cluster_found:
            break
    if not cluster_found:
        clusters[i] = [title1]

# Assign cluster labels to each job title
cluster_labels = {}
for cluster_id, cluster in clusters.items():
    for title in cluster:
        cluster_labels[title] = cluster_id

# Print the clusters and cluster labels
for cluster_id, cluster in clusters.items():
    print(f"Cluster {cluster_id + 1}: {', '.join(cluster)}")

```

```python
print("\nCluster Labels:")
for title in job_titles:
    print(f"{title}: Cluster {cluster_labels[title] + 1}")
```
